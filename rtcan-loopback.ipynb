```python
%matplotlib inline
import matplotlib
matplotlib.rcParams['figure.figsize'] = (16, 4)
matplotlib.rcParams.update({'font.size': 16})
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def loaddata(filename):
    return np.array([1e-3*float(line) for line in open(filename)])
```

```python
linux = loaddata('linux-linux-noload.txt')
xenomai = loaddata('xenomai-xenomai-rt-noload.txt')
```

```python
def showhist(data, title):
    plt.hist(data, bins=np.logspace(1, 5, 200))
    plt.gca().set_xscale("log")
    plt.gca().set_yscale("log")
    plt.gca().set_ylim([1e-1,1e7])
    plt.title(title)
    plt.xlabel('latency, us')
    plt.ylabel('events number')
    plt.grid(True)
    plt.show()
```

```python
showhist(linux, 'linux')
showhist(xenomai, 'xenomai')
```
